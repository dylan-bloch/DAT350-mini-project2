This is my mini project 2. For this project I am analyzing a lot of different snowboards and their effectiveness in riding in fresh powdery snow. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rawSnowboardDF = pd.read_csv("Snowboard2.csv")

rawSnowboardDF.head()
```

```python
brandGroup = rawSnowboardDF.groupby("Brand").size()

plt.rcParams["text.color"] = "black"
plt.rcParams["font.size"] = 10
plt.title("Brands")

plt.pie(brandGroup, labels=brandGroup.index,
        autopct = '%1.0f%%')
```

I cleared the brand output just because nearly every board had its own brand so the pie chart is very ugly. The labels and colors blend together, can't really get it to tell a story.
